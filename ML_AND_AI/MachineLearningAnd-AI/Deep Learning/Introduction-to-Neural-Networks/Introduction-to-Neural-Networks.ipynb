# 1. Activation Functions:
    ● Define the sigmoid function, a popular activation function used in neural networks.

```python
def sigmoid(x):
  num  = 1
  den = (1 + math.e**(-x))
  return num/den
```

```python
x = np.arange(-30, 30)
y = sigmoid(x)
plt.vlines(x=0, ymin=0, ymax=1, color='g',  linestyles='solid')
plt.hlines(y=0.5, xmin=-30, xmax=30, color='g',  linestyles='solid')
plt.plot(x,y)
```

    ● Define the tanh function, which is another common activation function that returns values between -1 and 1.

```python
def tanh(x):
  num = math.e**(x) - math.e**(-x)
  den = math.e**(x) + math.e**(-x)
  return num/den
```

```python
x = np.arange(-30, 30)
y = tanh(x)
plt.vlines(x=0, ymin=-1, ymax=+1, color='g')
plt.hlines(y=0, xmin=-30, xmax=30, color='g')
plt.plot(x, y)
```

# 2. Sample Data:
    ● Use the following sample data for training:
    x1=3, x2=2, yactual=17

```python
# input_feature1 | input_feature2 | output

x1 = 3  # input1
x2 = 2  # input2
ya = 17 # y_actual
```

# 3. Feed Forward:
    ● Initialize the random weights w1 and w2 between 1 and 10.

```python
# Phase I - Feed Forward
# 1.1 we need random weights
# 1.2 affine equations
# y_pred = x1*w1 + x2*w2

w1 = np.random.randint(1, 10)
w2 = np.random.randint(1, 10)
```

    ● Predict the output ypred using the equation ypred= x1 x w1 + x2 x w2

```python
# 1.2 Affine Equation
y_pred = x1*w1 + x2*w2
y_pred
```

# 4. Error Calculation:
    ● Compute the squared error as error = (yactual-ypred)*(yactual-ypred)

```python
error = (ya - y_pred)**2
```

# 5. Back Propagation:
    ● Compute the gradients of the error with respect to the weights.


```python
# Phase III - Back Propagation
# 3.1 Calculate derivates
# w1 = w1 - de/dw

dedw1 = 2*(ya-y_pred)*(-x1)
dedw2 = 2*(ya-y_pred)*(-x2)
```

    ● Update the weights using the computed gradients and a learning rate.

```python

# 3.2 Weightage Updation
w1 = w1 - dedw1
w2 = w2 - dedw2
```

# 6. Training:
    ● The above steps (Feed Forward to Back Propagation) are performed iteratively (20 times in this example) to refine the weights and minimize the error.

```python
lr = 0.01 #learnign rate
w1_list = []
w2_list = []
error_list = []
y_pred_list = []

for epoch in range(20):
  # Collect info in list
  w1_list.append(w1)
  w2_list.append(w2)

  # 1 Feed Forward
  # 1.1 Affin Equation
  y_pred = x1*w1 + x2*w2
  y_pred_list.append(y_pred)

  # 2 Error Calculation
  # 2.1 error
  error = (ya - y_pred)**2
  error_list.append(error)

  # 3 back Prop
  # 3.1 Calculate the derivates
  dedw1 = 2*(ya-y_pred)*(-x1)
  dedw2 = 2*(ya-y_pred)*(-x2)

  # 3.2 Update the weights
  w1 = w1 - lr * dedw1
  w2 = w2 - lr * dedw2
print(f"Value of y_predicted: {y_pred} & should be close to {ya}")
```
