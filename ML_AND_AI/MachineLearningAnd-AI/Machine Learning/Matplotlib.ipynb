# Description of the Dataset:
    This dataset contains these columns: id, date, price, bedrooms, bathrooms,
    sqft_living, sqft_lot, floors, waterfront, view, condition, grade, sqft_above,
    sqft_basement, yr_built, yr_renovated, zip code, lat, long, sqft_living15, sqft_lot15.
    Data Visualization


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("home_data.csv")
# sdata = sns.load_dataset("home_data")
data.head()
```

```python
data.info()
```

```python
sns.scatterplot(data = data, x="price", y="bedrooms", linewidth =0, alpha = 0.1,  hue='grade' )
```

# Python Correlation: Creating A Scatter Plot

```python
sns.scatterplot(data = data, x="price", y="grade", linewidth =0, alpha = 0.2 )
```

# Python Correlation: Creating A Regression Plot

```python
sns.regplot (data = data, x="price", y="grade")
```

```python
# change color effect
scatter_kws={"color" : "white", "edgecolor" : "grey", "alpha" : 0.4}
line_kws={"color" : "red", "linewidth" : 1}
sns.regplot (data = data, x="price", y="grade", scatter_kws=scatter_kws, line_kws=line_kws)
```

```python
sns.relplot(data=data, x="sqft_living", y="grade", hue="price")
```

# Python Correlation: Creating A Correlation Matrix

```python
data.corr()
```

# Python Correlation: Creating a Heat Map

```python
# Creating heat map
fig = plt.figure(figsize = (10,5))
sns.heatmap(data.corr())
```

```python
# Creating heat map
fig = plt.figure(figsize = (10,5))
sns.heatmap(data.corr(), linewidths=0.5, annot=True)
```

```python
# Creating heat map
fig = plt.figure(figsize = (11,10))
sns.heatmap(data.corr(method="spearman"), linewidths=0.5, annot=True)
```

# Python Correlation: Creating a Staircase Visual

```python
import numpy as np
fig = plt.figure(figsize = (11,5))
plt.style.use("default")
df_corr = data.corr()
mask = np.triu(np.ones_like(df_corr,dtype="bool"))
sns.heatmap(data.corr(method="spearman"), mask=mask, linewidth=1, annot=True,cmap="coolwarm")
```

```python
# sns.box(x=data["price"],y=data["sqft_above"])
fig = plt.figure(figsize = (11,5))

data.boxplot(by ='price', column =['sqft_above'], grid = False)
```

```python
filter_year = data["yr_built"]>=2013
data_aftr_2014 = data[filter_year]
sns.set_style("whitegrid")
sns.boxplot(x = 'price', y = 'sqft_above', data = data_aftr_2014, hue='yr_built' )
```

```python
# sns.box(x="price", y="sqft_above")
filter_year = data["yr_built"]>=2013
data_aftr_2014 = data[filter_year]
sns.set_style("whitegrid")
sns.boxplot(x = 'price', y = 'sqft_above', data = data_aftr_2014, hue='yr_built' )
```
