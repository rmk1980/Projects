# 16. Write a function to perform the following operation:
    Increase the grade of the house by 1 if the sqft of the house is greater than 0 and less than
    equal to 400. If the total sqft of the house is greater than 400, increase the grade of the house
    by 2.
    After creating the above function, which syntax will be using when using apply function to run
    the above created function in the dataset?
        A. data.apply(your_function_name(arg1))
        B. data.apply_fun(function_name())
        C. data.app(data)
        D. None of the above

```python
import pandas as pd
import numpy as np
h_data = pd.read_csv("housing.csv")

def increase_grade(grade, sqft):
    if (sqft > 0) and (sqft <=400):
        return grade + 1
    else:
        return grade + 2
    
h_data["grade"] = h_data.apply(lambda x: increase_grade(x["grade"], x["sqft_living"]), axis = 1)
h_data.head()
```

# 32. If the correlation between the columns bathrooms and sqft_living is 0.754665, what all interpretations can be made about the two columns?
    A. A positive correlation between the two columns
    B. The columns show perfect correlation.
    C. A strong negative correlation between the two columns
    D. No correlation between the two columns.

# Answer A

The correlation coefficient between the columns bathrooms and sqft_living is 0.7546651. This indicates that there is a strong positive correlation between the two columns. In other words, as the number of bathrooms increases, so does the square footage of living space. However, it is important to note that correlation does not imply causation. There may be other factors that influence both the number of bathrooms and the square footage of living space. For example, the number of bedrooms or the location of the house may also be important factors to consider when analyzing house prices.
